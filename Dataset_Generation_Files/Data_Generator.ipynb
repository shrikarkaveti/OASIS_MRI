```python
import pandas as pd
```

```python
mr = pd.read_csv("OASIS3_MR_json.csv")
mr = mr[mr["scan category"] == "T2w"]
mr_reduced = mr[["subject_id", "label"]]
mr_days = []

for i in range(len(mr_reduced)):
    mr_days.append(pd.to_numeric(mr_reduced.iat[i, 1][-4:]))

mr_reduced.insert(2, "days", mr_days)
mr_reduced_sorted = mr_reduced.sort_values(by = ["subject_id", "days"]).drop_duplicates().reset_index(drop = True)
mr_reduced_sorted.head(10)
```

```python
mr_list = []

for i in range(len(mr_reduced_sorted)):
    mr_list.append([mr_reduced_sorted.iat[i, 0], mr_reduced_sorted.iat[i, 1], pd.to_numeric(mr_reduced_sorted.iat[i, 1][-4:])])

print(len(mr_list))
print(mr_list[0])
```

```python
cdr = pd.read_csv("OASIS3_UDSb4_cdr.csv")
cdr_reduced = cdr[["OASISID", "OASIS_session_label", "days_to_visit", "age at visit", "MMSE", "memory", "orient", "judgment", "commun", "homehobb", "perscare", "CDRTOT"]]
cdr_reduced_sorted = cdr_reduced.sort_values(by = ["OASISID", "days_to_visit"]).reset_index(drop = True)
cdr_reduced_sorted.head(20)
```

```python
cdr_list = []

for i in range(len(cdr_reduced_sorted)):
    cdr_temp = []
    for j in range(len(cdr_reduced_sorted.columns)):
        cdr_temp.append(cdr_reduced_sorted.iat[i, j])
    cdr_list.append(cdr_temp)

print(len(cdr_list))
print(cdr_list[0])
```

```python
join_list = []
mr_index = 0
cdr_index = 0

while((mr_index < len(mr_list)) and ((cdr_index + 1) < len(cdr_list))):
    if (mr_list[mr_index][0] == cdr_list[cdr_index][0]):
        if (cdr_list[cdr_index + 1][2] > mr_list[mr_index][2]):
            temp_join_list = [mr_list[mr_index][1], cdr_list[cdr_index][1]]
            temp_join_list.extend(cdr_list[cdr_index][2:])
            join_list.append(temp_join_list)
            mr_index = mr_index + 1
        else:
            cdr_index = cdr_index + 1
    else:
        if (mr_list[mr_index][0] > cdr_list[cdr_index][0]):
            cdr_index = cdr_index + 1
        else:
            mr_index = mr_index + 1
```

```python
for i in join_list:
    print(i)
```

```python
print(len(join_list))
```

```python
cdr_count = [0, 0, 0, 0]
csv = open("download_data.csv", "w")

for i in join_list:
    # Create CSV
    csv.write("{}\n".format(i[0]))

    if (i[-1] == 0):
        cdr_count[0] = cdr_count[0] + 1
    elif (i[-1] == 0.5):
        cdr_count[1] = cdr_count[1] + 1
    elif (i[-1] == 1):
        cdr_count[2] = cdr_count[2] + 1
    elif (i[-1] == 2):
        cdr_count[3] = cdr_count[3] + 1
    print(i)

csv.close()
```

```python
# Write csv file for Train and Test
data_gen_csv = open("data_gen_csv.csv", "w")
header = ["MRI_ID", "OAS_Session_ID", "days_to_visit", "age_at_visit", "MMSE", "memory", "orient", "judgment", "commun", "homehobb", "perscare", "CDRTOT"]
csv_join = [header]
csv_join.extend(join_list)

for i in csv_join:
    for j in range(len(i)):
        if (j < len(i) - 1):
            data_gen_csv.write("{},".format(i[j]))
        else:
            data_gen_csv.write("{}\n".format(i[j]))

data_gen_csv.close()
```
